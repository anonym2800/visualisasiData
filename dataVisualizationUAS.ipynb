```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Mengambil data dari resource
data = pd.read_csv('/content/kebutuhan_per_bulan_2.csv')
data
```

```python
# Filtering propinsi
data_filtered = data [data['Propinsi']=='MALUKU']
data_filtered
```

```python
# TABEL PERTAMA
x = data_filtered['tahun']
y = data_filtered['kebutuhan']

plt.scatter(x, y)

#add label
plt.xlabel("Tahun", fontsize = 13)
plt.ylabel("Kebutuhan" , fontsize = 13)
plt.title('Kebutuhan per bulan pada Maluku', fontweight ="bold",)
plt.show()
```

```python
# Filtering Propinsi
data_filtered_jabar = data [data['Propinsi']=='JAWA BARAT']
data_filtered_jabar
```

```python
# TABEL KEDUA
y = data_filtered_jabar["kebutuhan"]
x = data_filtered_jabar["tahun"]

plt.scatter(x, y, color='red')

#add label
plt.xlabel("Tahun", fontsize = 13)
plt.ylabel("Kebutuhan" , fontsize = 13)
plt.title('Kebutuhan per bulan pada Jawa Barat', fontweight ="bold",)
plt.show()
```

```python
# TABEL KETIGA
data3 = pd.DataFrame(data)
data3_year = data3 [data3['tahun']== 2014]

top10 = data3_year.head(10)

x = top10.Propinsi
y = top10.kebutuhan

plt.scatter(x,y)
plt.xlabel("Propinsi", fontsize = 13)
plt.ylabel("Kebutuhan" , fontsize = 13)
plt.title('top10 Propinsi pada tahun 2014', fontweight ="bold",)
plt.xticks(rotation=90)
plt.show()
```

```python
dataProduk = pd.read_csv('/content/produk-new.csv')
```

```python
# TABEL EMPAT
produk = pd.DataFrame(dataProduk)
produk_sort = produk.sort_values(by='Price', ascending=True)
produk_clean = produk_sort.loc[produk['Price'] != 0].head(10)

x = produk_clean['Price']
y = produk_clean['productName']

plt.xlabel("Harga", fontsize = 13)
plt.ylabel("Produk" , fontsize = 13)
plt.title('top10 Produk dengan Harga Murah', fontweight ="bold",)

plt.scatter(x,y)
plt.xticks(rotation=90)
plt.show()
```

```python
data_transaksi = pd.read_csv("/content/transaksi2.csv" , names=['Date', 'ID', 'Quantity', 'Price'])
data_transaksi
```

```python
# TABEL LIMA
data_transaksi['Date'] = pd.to_datetime(data_transaksi['Date'])
data_transaksi['year'] = data_transaksi['Date'].dt.year
data_transaksi['month'] = data_transaksi['Date'].dt.month_name()

data_year = data_transaksi[data_transaksi['year'] == 2017]
data_year

plt.scatter(data_year['month'], data_year['Quantity'])

plt.xlabel('Month')
plt.ylabel('Quantity')
plt.title('Transaksi pada tahun 2017', fontweight ="bold",)
plt.xticks(rotation=90)
plt.show()
```

```python
#Tabel enam
data_transaksi['Date'] = pd.to_datetime(data_transaksi['Date'])
data_transaksi['year'] = data_transaksi['Date'].dt.year
data_transaksi['month'] = data_transaksi['Date'].dt.month_name()

data_year = data_transaksi[data_transaksi['year'] == 2017]
data_year

data_may = data_year [data_year['month']=='May']
data_may

plt.scatter(data_may['Quantity'], data_may['Date'])

plt.xlabel('Quantity')
plt.ylabel('Date')
plt.title('Transaksi pada Bulan May 2017', fontweight ="bold",)
plt.show()
```
